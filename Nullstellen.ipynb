Fixpunktiteration

$f(x) = x^{3} -x +0,3$

$F(x) = x^{3}+0,3$

```python
startwerte = [-1, 0, 1]
for wert in startwerte:
    i = 0
    x = wert
    print("Suche für Startwert: " , x)
    while i<4:
        i = i +1
        x= x ** 3 +0.3
        print(x)
```

Sekantenverfahren

für $f(x) = x^{2} -2$

$$\large x_{n+1} = x_n - \frac{f(x_n)} {f'(x_n)}$$

ersetzte $f'(x_n)$ durch $∆_n$

$$ \Large ∆_n = \frac{f({x}_{n}) - f(x_{n-1})} {x_n - x_{n-1}}$$

```python
i = 0
x0 = -99989893453458
x1 = 1.4
print("Suche für Startwert: " , x0, "und", x1)
while i<5:
    i = i +1
    f1 = (x1 ** 2) -2
    f0 = (x0 ** 2) -2
    #print(f0)
    #print(f1)
    delta = (f1 - f0) / (x1 - x0)
    x0 = x1
    x1= x1 - (x1 ** 2 -2)/delta
    print(x1)
```
