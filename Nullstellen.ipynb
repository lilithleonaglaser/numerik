Bisektionsverfahren

$\large f(x) = e^{x²/10} -2$


```python
import math

i = 0
a0 = 2
b0 = 3
def f(x):
    return math.exp((x ** 2)/10) - 2
print("Suche für Startwert: " , a0, "und", b0)
while i<8:
    i = i +1
    fa = f(a0)
    fb = f(b0)
    print("a: ", a0)
    print("b: ",b0)
    #print("fa: ", fa)
    #print("fb: ",fb)

    x = 0.5 * (a0+b0)
    fx = f(x)


    if (fa * fx) < 0:
        b0 = x
    elif (fx * fb) < 0:
        a0 = x
    else:
        print("Nullstelle gefunden ", x)

    print("x",i,": ", x)
    print("fx: ",fx)
```

Fixpunktiteration

$f(x) = x^{3} -x +0,3$

$F(x) = x^{3}+0,3$

```python
startwerte = [-1, 0, 1]
for wert in startwerte:
    i = 0
    x = wert
    print("Suche für Startwert: " , x)
    while i<4:
        i = i +1
        x= x ** 3 +0.3
        print(x)
```

Newton-Verfahren

$\large f(x) = e^{x²/10} -2$


```python
i = 0
x = 2
def f(x):
    return math.exp((x ** 2)/10) - 2
def f_strich(x):
    return (x * math.exp((x ** 2)/10)) / 5
print("Suche für Startwert: " , x)
while i<8:
    i = i +1
    x = x - (f(x)/f_strich(x))
    print(x)
```
